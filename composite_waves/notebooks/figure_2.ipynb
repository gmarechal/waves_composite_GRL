```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
pol = 'cyclonic' # polarization
path_data = '../data/theory' + "/"
path_output = '../figures/'
```

### Load data

```python
ds_swell = xr.open_dataset(path_data + 'composite_swell.nc')
ds_ws = xr.open_dataset(path_data + 'composite_windsea.nc')
ds_swell_theo = xr.open_dataset(path_data + 'composite_swell_theo.nc')
ds_ws_theo = xr.open_dataset(path_data + 'composite_windsea_theo.nc')

data_swell = ds_swell.composite_swell
data_ws = ds_ws.composite_windsea
data_swell_theo = ds_swell_theo.composite_swell
data_ws_theo = ds_ws_theo.composite_windsea
```

### Plot Circles

```python
theta = np.linspace(0, 2*np.pi, 100)
x_circle = np.cos(theta)
y_circle = np.sin(theta)

x_circle_small = np.cos(theta)/2.5
y_circle_small = np.sin(theta)/2.5


arrow_length = 0.8  # Adjust length as needed
x_start = 0
y_start = 0
theta_rot = 0
x_end = x_start + arrow_length * np.cos(theta_rot*(+1))
y_end = y_start + arrow_length * np.sin(theta_rot*(+1))

props1 = dict(boxstyle='round', facecolor='w', alpha=1)
n_quiver = 15
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, sharex = True, sharey = True)
ax = axes[0][0]
p1 = ax.pcolormesh(data_swell.x, data_swell.y, data_swell, vmin = -0.06, vmax = 0.06, cmap = 'RdBu_r')
ax.plot([0, 0], [-1, 1], color = 'k')
ax.plot(x_circle_small, y_circle_small, color = 'k', linewidth = 3)
ax.plot(x_circle_small, y_circle_small, color = 'w', linewidth = 1)
ax.plot(x_circle, y_circle, color = 'k', linewidth = 3)
ax.plot(x_circle, y_circle, color = 'w', linewidth = 1, linestyle = '--')
ax.set_xticks([-1, -.4, 0, .4, 1], ['-2.5$R_{max}$', '-$R_{max}$', '0', '$R_{max}$', '2.5$R_{max}$'], rotation = 90)
ax.set_yticks([-1, -.4, 0, .4, 1], ['-2.5$R_{max}$', '-$R_{max}$', '0', '$R_{max}$', '2.5$R_{max}$'])

# ax.set_title('Model Swell')
ax.set_aspect('equal', 'box')
ax.text(0.05, 0.9, '(a) Model Swell',
     transform = ax.transAxes, bbox = props1, fontsize = 9)

ax = axes[0][1]
ax.pcolormesh(data_swell.x, data_swell.y, data_swell_theo, vmin = -0.06, vmax = 0.06, cmap = 'RdBu_r')
ax.set_aspect('equal', 'box')
ax.plot([0, 0], [-1, 1], color = 'r')
ax.plot(x_circle_small, y_circle_small, color = 'k', linewidth = 3)
ax.plot(x_circle_small, y_circle_small, color = 'w', linewidth = 1)
ax.plot(x_circle, y_circle, color = 'k', linewidth = 3)
ax.plot(x_circle, y_circle, color = 'w', linewidth = 1, linestyle = '--')
ax.text(0.05, 0.9, '(b) Theory Swell',
     transform = ax.transAxes, bbox = props1, fontsize = 9)

ax = axes[1][0]
ax.pcolormesh(data_swell.x, data_swell.y, data_ws, vmin = -0.06, vmax = 0.06, cmap = 'RdBu_r')
ax.set_aspect('equal', 'box')
ax.plot([0, 0], [-1, 1], color = 'k')
ax.plot(x_circle_small, y_circle_small, color = 'k', linewidth = 3)
ax.plot(x_circle_small, y_circle_small, color = 'w', linewidth = 1)
ax.plot(x_circle, y_circle, color = 'k', linewidth = 3)
ax.plot(x_circle, y_circle, color = 'w', linewidth = 1, linestyle = '--')
ax.set_xticks([-1, -.4, 0, .4, 1], ['-2.5$R_{max}$', '-$R_{max}$', '0', '$R_{max}$', '2.5$R_{max}$'], rotation = 90)
ax.set_yticks([-1, -.4, 0, .4, 1], ['-2.5$R_{max}$', '-$R_{max}$', '0', '$R_{max}$', '2.5$R_{max}$'])
ax.text(0.05, 0.9, '(c) Model Wind Sea',
     transform = ax.transAxes, bbox = props1, fontsize = 9)

ax = axes[1][1]
ax.pcolormesh(data_swell.x, data_swell.y, data_ws_theo, vmin = -0.06, vmax = 0.06, cmap = 'RdBu_r')
ax.set_aspect('equal', 'box')
ax_inset = fig.add_axes([.82, .6, .15, .357])
ax_inset.plot(data_swell[:, 50], data_swell.y, color = 'k', label = 'Model')
ax_inset.plot(data_swell_theo[:, 50], data_swell.y, color = 'r', label = 'Theory')
ax_inset.set_xlim([-0.07, 0.07])
ax_inset.set_yticks([])
ax_inset.legend(bbox_to_anchor = (.7, 1))
ax_inset.set_xticks([])
# ax_inset.set_xticklabels(['-0.05', '0', '0.05'], rotation = 30)
ax_inset.set_ylim([-1, 1])
ax_inset.text(0.1, 0.9, '(e)',
     transform = ax_inset.transAxes, bbox = props1, fontsize = 9)

ax.text(0.05, 0.9, '(d) Theory Wind Sea',
     transform = ax.transAxes, bbox = props1, fontsize = 9)
ax.plot([0, 0], [-1, 1], color = 'r')
ax.plot(x_circle_small, y_circle_small, color = 'k', linewidth = 3)
ax.plot(x_circle_small, y_circle_small, color = 'w', linewidth = 1)
ax.plot(x_circle, y_circle, color = 'k', linewidth = 3)
ax.plot(x_circle, y_circle, color = 'w', linewidth = 1, linestyle = '--')
ax.set_xticks([-1, -.4, 0, .4, 1], ['-2.5$R_{max}$', '-$R_{max}$', '0', '$R_{max}$', '2.5$R_{max}$'], rotation = 90)
ax.set_yticks([-1, -.4, 0, .4, 1], ['-2.5$R_{max}$', '-$R_{max}$', '0', '$R_{max}$', '2.5$R_{max}$'])



ax_inset = fig.add_axes([.82, .18, .15, .357])
ax_inset.plot(data_ws[:, 50], data_ws.y, color = 'k', label = 'Model')
ax_inset.plot(data_ws_theo[:, 50], data_ws_theo.y, color = 'r', label = 'Theory')
ax_inset.set_yticks([])
ax_inset.set_ylim([-1, 1])

ax_inset.set_xlim([-0.07, 0.07])
ax_inset.set_xlabel('$H_s$ anomaly [m]')
ax_inset.set_xticks([-0.05, 0, 0.05])
ax_inset.set_xticklabels(['-0.05', '0', '0.05'], rotation = 30)
ax_inset.text(0.1, 0.9, '(f)',
     transform = ax_inset.transAxes, bbox = props1, fontsize = 9)
cax = fig.add_axes([.3, 1.08, 0.4, 0.02])
plt.subplots_adjust(wspace=-.8, hspace=0)

plt.colorbar(p1, cax = cax, orientation = 'horizontal', label = '$H_s$ Anomaly [m]')
plt.tight_layout()
plt.savefig(path_output + 'Fig2.png', dpi = 300, bbox_inches = 'tight')
```
